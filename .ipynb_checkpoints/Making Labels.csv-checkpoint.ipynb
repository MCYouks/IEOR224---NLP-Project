```python
clustlistcsv=pd.DataFrame(clustlistcsv)
stockscsv=pd.DataFrame(os.listdir(path))

clustlistfinal = pd.merge(clustlistcsv,stockscsv)
clustlistfinal.columns = ['Stocks']

clustlistfinallist = clustlistfinal['Stocks'].values.tolist()
#clustlistfinallist = list(clustlistfinal[,])
print(clustlistfinallist)
len(clustlistfinallist)
```

```python
clustpath='/Users/Jillian/Desktop/UCB MEng/Spring 2018/INDENG 224/224-p/IEOR224---NLP-Project/StocksClustered'
```

```python
labels=[]

for i in range(0, len(clustlistfinallist)):
    #for j in clustlistfinallist:
    if clustlistfinallist[i] != ".DS_Store":
        df = pd.read_csv(os.path.join(clustpath, clustlistfinallist[i]))
        
        labels.append(list(df['Label']))
        
        labelsdf = pd.DataFrame(labels).transpose()
        
        
labelsdf        

```
