```python
## producing stock labels from all stocks, with labels

import os
import pandas as pd
import numpy as np

dfList = []
path = '/Users/Jillian/Desktop/UCB MEng/Spring 2018/INDENG 224/224-p/IEOR224---NLP-Project/Stocks/' 
newpath = '/Users/Jillian/Desktop/UCB MEng/Spring 2018/INDENG 224/224-p/IEOR224---NLP-Project/StocksNew/'

for fn in os.listdir(path): 
    if fn != ".DS_Store":
        # Absolute file path
        file = os.path.join(path, fn)
        if os.path.isfile(file): 
            #file.encode('utf-8').strip()
            df = pd.read_csv(file)
          
            adj_close=list(df['Adj Close'])
            adj_closeadj=[]

            for i in range(0,len(adj_close)):
                if adj_close[i]=='null':
                    adj_closeadj.append(0.0)
                else:
                    adj_closeadj.append(adj_close[i])
                    
                adj_closeadj[i]=float(adj_closeadj[i])
            
            px_change=[0,]
            
            for i in range(0,len(adj_closeadj)-1):
                j = i + 1
                px_change.append(adj_closeadj[j]-adj_closeadj[i])

                #print(px_change)
                    
            # we then add the series to the dataframe, which holds our parsed CSV file
            df['Percent Change'] = px_change
            # save the dataframe to CSV
            #df.to_csv('path/to/file.csv', sep='\t')
            df['Label'] = (df['Percent Change'] > 0).astype('int')

            df.to_csv(os.path.join(newpath, fn))

        #dfList.append(df) 

#appended_data = pd.concat(dfList)
#appended_data.to_csv(os.path.join(newpath, fn))
```
