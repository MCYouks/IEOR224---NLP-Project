```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from datetime import datetime

```

```python
headlines=pd.read_csv('224 Proj Headlines For Stocks.csv')
headlines.set_index('Date', drop=True, inplace=True)
headlines.head()
```

```python
labels=pd.read_csv('Labels.csv')
labels.set_index('Date', drop=True, inplace=True)
labels.head()
```

```python
labels.index = pd.to_datetime(labels.index)
labels.head()
```

```python
hll=headlines.index.values.tolist()
hll
```

```python
dates =[]

for i in range(0,len(hll)):
    datetimeobject = datetime.strptime(hll[i],'%d/%m/%Y')
    newformat = datetimeobject.strftime('%Y-%m-%d')
    
    dates.append(newformat)

headlines.index = dates
headlines.head()
#headlines.index = pd.to_datetime(headlines.index)
#headlines
```

```python
stock=labels.iloc[:,1:2]
data = pd.concat([stock, headlines], axis=1, join='inner')
train = data[data.index < '2016-03-31']
print(train.iloc[0,1:27])
```

```python
len(data)
```

```python
totalpredacc = []

for i in range(0, len(labels)):
  
    stock=labels.iloc[:,i:i+1]
    data = pd.concat([stock, headlines], axis=1, join='inner')

    data.fillna(0, inplace=True)

    #using dates btwn 2016-03-31 and 2018-03-31 as test, rest as train
    train = data[data.index < '2016-03-31']
    test = data[data.index > '2016-03-31']

    col=list(test.columns)

    print('Testing on Stock',i+1,',', col[0])

    trainheadlines = []

    for row in range(0,len(train.index)):
        trainheadlines.append(' '.join(str(x) for x in train.iloc[row,1:len(train.columns)]))

    basicvectorizer = CountVectorizer()
    basictrain = basicvectorizer.fit_transform(trainheadlines)
    #print(basictrain.shape)

    basicmodel = LogisticRegression()
    basicmodel = basicmodel.fit(basictrain, train.iloc[:,0])

    testheadlines = []
    for row in range(0,len(test.index)):
        testheadlines.append(' '.join(str(x) for x in test.iloc[row,1:len(train.columns)]))

    basictest = basicvectorizer.transform(testheadlines)

    predictions = basicmodel.predict(basictest)

    print("")


    predtab = pd.crosstab(test[col[0]], predictions, rownames=["Actual"], colnames=["Predicted ->"],dropna=False)

    print('Confusion Matrix for', col[0], 'Stock :')

    zero = pd.DataFrame([[0,0]], columns=predtab.columns)

    if len(predtab.index) < 2:
        predtab=predtab.append(zero, ignore_index=True)

    print(predtab)

    predacc = (predtab.iloc[0][0] + predtab.iloc[1][1]) / (predtab.values.sum())
    print("")
    print('Prediction Accuracy for', col[0], 'Stock :',predacc)

    print("")
    print("--------------------------------------")
    print("")
    totalpredacc.append(predacc)
        
totalpredacc        
```

```python
modelacc = pd.DataFrame()
modelacc['Stock'] = list(labels.columns)
modelacc['Prediction Accuracy'] = totalpredacc
modelacc
```

### Prediction with n-gram 2


```python
#This is where the n-gram model comes in.
#In this model, n represents the length of a sequence of words to be counted.
#This means our bag-of-words model was the same as an n-gram model where n = 1.
#So now, let's see what happens when we run an n-gram model where n = 2.

#Below, we'll create a new CountVectorizer with the n-gram parameter set to 2 instead of the default value of 1.
```

```python
advancedvectorizer = CountVectorizer(ngram_range=(2,2))
advancedtrain = advancedvectorizer.fit_transform(trainheadlines)
print(advancedtrain.shape)

#This time we have 1115019 unique variables representing two-word combinations!

```

```python
totalpredacc2 = []

for i in range(0, len(labels)):
  
    stock=labels.iloc[:,i:i+1]
    data = pd.concat([stock, headlines], axis=1, join='inner')

    data.fillna(0, inplace=True)

    #using dates btwn 2016-03-31 and 2018-03-31 as test, rest as train
    train = data[data.index < '2016-03-31']
    test = data[data.index > '2016-03-31']

    col=list(test.columns)

    print('Testing N-Gram 2 on Stock',i+1,',', col[0])

    trainheadlines = []

    for row in range(0,len(train.index)):
        trainheadlines.append(' '.join(str(x) for x in train.iloc[row,1:len(train.columns)]))

    advancedvectorizer = CountVectorizer(ngram_range=(2,2))
    advancedtrain = advancedvectorizer.fit_transform(trainheadlines)

    advancedmodel = LogisticRegression()
    advancedmodel = advancedmodel.fit(advancedtrain, train.iloc[:,0])

    testheadlines = []
    for row in range(0,len(test.index)):
        testheadlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
    advancedtest = advancedvectorizer.transform(testheadlines)
    advpredictions = advancedmodel.predict(advancedtest)

    print("")


    advpredtab = pd.crosstab(test[col[0]], advpredictions, rownames=["Actual"], colnames=["Predicted ->"],dropna=False)

    print('Confusion Matrix for', col[0], 'Stock :')

    zero = pd.DataFrame([[0,0]], columns=advpredtab.columns)

    if len(advpredtab.index) < 2:
        advpredtab=advpredtab.append(zero, ignore_index=True)

    print(advpredtab)

    advpredacc = (advpredtab.iloc[0][0] + advpredtab.iloc[1][1]) / (advpredtab.values.sum())
    print("")
    print('Prediction Accuracy for N-Gram=2 for', col[0], 'Stock :',advpredacc)

    print("")
    print("--------------------------------------")
    print("")
    totalpredacc2.append(advpredacc)
        
totalpredacc2        
```

```python
modelacc2 = pd.DataFrame()
modelacc2['Stock'] = list(labels.columns)
modelacc2['Prediction Accuracy'] = totalpredacc2
modelacc2
```
